### Part 5: Calculate the average number of toppings per burrito.

Note: Let's ignore the `quantity` column to simplify this task.

Hint: Think carefully about the easiest way to count the number of toppings!


```python
chipotle['choice_description']
```

```python
def burrito_finder(x):
    if 'burrito' in str.lower(x):
        return 1
    else:
        return 0
    
chipotle['burrito_flag'] = chipotle['item_name'].apply(burrito_finder)
```

```python
chipotle[chipotle[ 'burrito_flag'] == 1 ]

def topping_counter(x):
    return x.count(',') + 1

chipotle['topping_count'] = chipotle[chipotle['burrito_flag'] == 1]['choice_description'].apply(topping_counter)

chipotle.head(20)

chipotle['topping_count'].mean()
```
